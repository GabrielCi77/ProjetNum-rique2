#### Réponse 1

$ \alpha $ est le taux de reproduction des proies

$ \beta $ est le nombre de proies mangées par un prédateur pendant un petit intervalle de temps dt

$ \gamma $ est le taux de mort naturelle des prédateurs en l’absence de proies

$ \delta $ est la dérivée partielle du taux de mortalité des prédateurs par rapport à la quantité de proies

On définit f par : f(t,(x1,x2)) = (x1($\alpha$-$\beta$ x2), -x2($\gamma$-$\delta$ x1))

Pour simplifier les notations, on écrira seulement la dépendance de f en (x1,x2) et pas en t

Alors,

$f(x1,x2) = 0 \Leftrightarrow \left\{
    \begin{array}{ll}
    x1 = 0\\
    x2 = 0\\
    \end{array}
    \right.
   ou \left\{
   \begin{array}{ll}
   x1 = \frac{\gamma}{\delta}\\
   x2 = \frac{\alpha}{\beta}\\
   \end{array}
   \right.$
   
Il y a donc deux points d'équilibre : (0,0) et $\overline{x} = (\frac{\gamma}{\delta},\frac{\alpha}{\beta})$

La jacobienne de f vaut : 

$J_f (x1,x2) = \begin{pmatrix} \alpha - \beta x2 & -\beta x1 \\ \delta x2 & \delta x1 - \gamma \end{pmatrix} $

$J_f (0,0) = \begin{pmatrix} \alpha & 0 \\ 0 & -\gamma \end{pmatrix}$

Le polynôme caractéristique vaut $\chi (J_f (0,0)) = (X - \alpha) (X + \gamma)$

$\alpha > 0$ et $\alpha$ est valeur propre donc (0,0) est instable

$J_f (\overline{x}) = \begin{pmatrix} 0 & -\frac{\beta \gamma}{\delta} \\ \frac{\alpha \delta}{\beta} & 0 \end{pmatrix}$

Le polynôme caractéristique vaut $\chi (J_f (\overline{x})) = X^2 + \alpha \gamma = (X - i \sqrt{\alpha \gamma}) (X + i \sqrt{\alpha \gamma})$

Les valeurs propres sont $i \sqrt{\alpha \gamma}$ et $-i \sqrt{\alpha \gamma}$
Elles sont toutes deux de partie réelle nulle donc on ne peut pas conclure quant à la stabilité de $\overline{x} $

$y(t) = (0, x_2(t_2) e^{\gamma (t_2 - t)})$

Soit x $\in S_f (0,x_0)$ avec $x_0 \in \mathbb{R}_{>0}$

Soit I l'intervalle de définition de x

Soit $t_m^+$ = sup I

Supposons $t_m^+$ fini

D'après le théorème du domaine maximal, (t,x(t)) tend vers la frontière du domaine de définition de f ou diverge
Comme f est continue et définie sur $\mathbb{R}_{>0} \times \mathbb{R}_{>0}$,

lim ||x(t)|| = +$\infty$

Avec la norme 1, on a lim (x1(t)+x2(t)) = +$\infty$

Or H(x1,x2)>= min($\beta$,$\delta$)(x1+x2) - ln(max($\alpha$,$\gamma$)) - ln(x1+x2)

Lorsque ||(x1,x2)|| tend vers $+\infty$, x1+x2 tend vers $+\infty$, donc le terme de droite tend vers $+\infty$ et donc H(x1,x2) tend aussi vers $+\infty$

Donc H(x) tend vers $+\infty$ lorsque t tend vers $t_m^+$

C'est absurde car H est constante donc ne diverge pas

Donc $t_m^+$ n'est pas fini, $t_m^+$=$+\infty$

De même, inf I = $-\infty$

Ainsi, x est définie sur $\mathbb{R}$

Posons G définie par G(x1,x2) = H(x1,x2) - H($\overline{x}$)

G(x1,x2) = $\delta x1 - \gamma ln(x1) + \beta x2 - \alpha ln(x2) - \gamma (1-ln(\frac{\gamma}{\delta})) - \alpha (1-ln(\frac{\alpha}{\beta}))$

Par étude de fonction, il vient que 

$\forall x \in \mathbb{R}_{>0} \times \mathbb{R}_{>0}\setminus\lbrace{\overline{x}}\rbrace, G(x)>0$

et G($\overline{x}$) = 0

De plus, 

$\forall x \in \mathbb{R}_{>0} \times \mathbb{R}_{>0}$, <$\nabla$ G(x),f(x)> = <$\nabla$ H(x),f(x)> = $\frac{d}{dt}$H(x(t)) = 0

Donc G est une fonction de Lyupanov

Par le théorème de Lyupanov, $\overline{x}$ est stable.

```python
def display_contour(f, x, y, levels):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    fig, ax = plt.subplots()
    contour_set = plt.contour(
        X, Y, Z, colors="grey", linestyles="dashed", 
        levels=levels 
    )
    ax.clabel(contour_set)
    plt.grid(True)
    plt.xlabel("$x_1$") 
    plt.ylabel("$x_2$")
    plt.gca().set_aspect("equal")
```

```python
import math
def M(x,y):
    a, b, c, d = 1, 2, 3, 4
    n, p = np.shape(x)
    Z = np.zeros((n,p))
    for i in range(n):
        for j in range(p):
            Z[i][j]= d*x[i,j]-c*math.log(x[i,j])+b*y[i,j]-a*math.log(y[i,j])
    return Z
```

```python

display_contour(
    M, 
    x = np.linspace(0, 3.0,100)[1:], #on exclut 0 pour appliquer le logarithme sans problème
    y = np.linspace(0, 3.0,100)[1:], 
    levels=10 
)
```

```python
def solve_euler_explicit(f, x0, dt, t0, tf):
    ts, xs = [t0], [x0]
    t = t0
    x = x0
    while t < tf:
        t_next, x_next = t + dt, x + dt*f(t,x)
        ts = np.vstack([ts,t_next])
        xs = np.vstack([xs,x_next])
        t, x = t_next, x_next
    return ts, xs.T

```

```python
def f(t,x):
    return x
```

On va choisir $x_0 = \begin{pmatrix} 1 \\ 1 \end{pmatrix}$,

$\frac{dx}{dt} = x$, t0 = 0 et tf = 1

La solution théorique de cette équation est: 

$x(t) = exp(t I_2) x_0 = e^t x_0$

On devrait donc trouver $x(1) = \begin{pmatrix} e \\ e \end{pmatrix}$

Pour vérifier la convergence, on trace la valeur de $x(1)$ obtenue par le schéma d'Euler explicite avec un pas de temps de $\frac{1}{n}$ en fonction de $n$

```python
x0 = np.array([1,1])

X = [n for n in range(1,500)]
Y = [solve_euler_explicit(f, x0, 1/n, 0, 1)[1][0][-1] for n in range(1,500)]

plt.plot(X,Y)
plt.show()
```

On observe effectivement une convergence vers la valeur de e, soit environ 2.7

Afin de déterminer l'ordre de convergence, on va calculer, pour différents pas de temps dt, l'erreur maximale obtenue avec le schéma explicite.
Ensuite, on tracera le logarithme népérien de l'erreur maximale en fonction du logaritkme népérien du pas de temps.
L'ordre de convergence est le plus petit entier p tel qu'il existe une constante $c_s$ telle que :
$ln(\eta_j) \leq c_s + p*ln(\Delta t_j)$

```python
pas = np.linspace(0,1,100)[1:]
erreur_max = []
logpas=[]


for dt in pas:
    T, x = solve_euler_explicit(f, x0, dt, 0, 10)
    erreur = [(x[0][j+2]-x[0][j+1] - dt*f(T[j+1],x[0][j+1]))/dt for j in range(1,len(T)-2)]#erreur[j] est l'erreur êta(j+1)
    erreur_max.append(np.log(max(erreur)))# on retient le logarithme de l'erreur maximale pour le pas dt
    logpas.append(np.log(dt))

plt.plot(logpas,erreur_max)
plt.show()
```

On suppose que $f$ est lipschitzienne


```python
def solve_euler_implicit(f,x0,dt,t0,tf,itermax=100):
    t=[t0];x=[x0]
    i=1;
    while t0+i*dt < tf :
        t.append(t0+i*dt)
        def F_j (a):
            return x[i-1] + np.multiply(dt,(f(t[i],a)))
        k=0
        x_0=F_j(x[i-1])
        x_k=x_0
        x_k1=F_j(x_k)
        while np.linalg.norm((x_k1-x_k)/x_0) > 10**-5:
            if k < itermax :
                x_k=x_k1
                x_k1=F_j(x_k1)
                k+=1
            else :
                raise Exception("Trop d'itérations")
        x.append(x_k1)
        i+=1
    return t,x
```

```python
def f(t,x):
    return [float(x[1]),float(-x[1]-10*(x[0]))]
X,Y=solve_euler_implicit(f,[1.,1.],10**-3,0,20)
plt.plot(X,Y)
plt.title('Evolution angulaire et en vitesse du pendule')
plt.show()
```

On a bien ici une évolution convergente vers 0, en accord avec ce qu'on connait des pendules avec frottement, donc nous avons un programme qui semble bien converger

#### Réponse 10

$\frac{d}{dt} (H(x(t))-H_0) = \delta \frac{dx_1}{dt} - \frac{\gamma}{x_1}\frac{dx_1}{dt} + \beta \frac{dx_2}{dt} - \frac{\alpha}{x_2} \frac{dx_2}{dt}$

= $[(\frac{\gamma}{x_1} - \delta) u_1(x_1,x_2) + (\frac{\alpha}{x_2} - \beta) u_2 (x_1,x_2)] (H(x_1,x_2) - H_0)$

D'autre part,

$\nabla H(x) = \begin{pmatrix} \delta - \frac{\gamma}{x_1} & \beta - \frac{\alpha}{x_2} \end{pmatrix}$

Donc

$||\nabla H(x)||^2 = (\delta - \frac{\gamma}{x_1})^2 + (\beta - \frac{\alpha}{x_2})^2$

Ainsi, pour  $k \in \mathbb{N}$, en choisissant :

$u_1(x1,x2) = k (\delta - \frac{\gamma}{x_1})$

$u_2(x1,x2) = k (\beta - \frac{\alpha}{x_2})$

$u(x) = k \nabla H(x)$

On obtient : 

$\frac{d}{dt} (H(x(t))-H_0) = -k ||\nabla H(x)||^2 (H(x(t))-H_0)$


D'après l'énoncé, $x \rightarrow \| x - \overline{x} \|$ admet un minorant $k>0$ sur $\mathbb{R}^2$

De plus,  $x\rightarrow \| \nabla H (x)\| $ est continue

De plus, $\nabla H(x)$ s'annule seulement en $\overline{x}$

Donc la fonction $x\rightarrow \| \nabla H (x)\| $  admet un minorant $m>0$ sur $\mathbb{R}^2$ ,

$\forall t \in \mathbb{R}_+ , \| \nabla H (x(t))\| \geq m$


Posons $g(t) = (H(x(t)) - H_0) e^{km^2 t}$

Alors, $g'(t)=[\frac{d}{dt} (H(x(t))-H_0) + km^2 (H(x(t))-H_0)] e^{km^2t}$

Or $\frac{d}{dt} (H(x(t))-H_0) + km^2 (H(x(t))-H_0) \leq \frac{d}{dt} (H(x(t))-H_0) + k\| \nabla H (x(t))\|^2 (H(x(t))-H_0) = 0$

Donc $g'(t)\leq 0$

Donc

$\forall t \in \mathbb{R}, g(t) \leq g(0) = H(x(0))-H_0$

$\forall t \in \mathbb{R}, H(x(t))-H_0 \leq [H(x(0))-H_0]e^{-km^2t}$

Ainsi, $H(x)$ converge exponentiellement vers $H_0$ lorsque $t$ tend vers l'infini

Ici on est passé de k=10 à k=100 et cela fausse les résultats. le schéma n'est pas convergent
k est la vitesse de convergence de H. 

On pose :
$\tilde{f}(x) = (x_1(\alpha-\beta x_2)-u_1(x_1,x_2)(H(x_1,x_2)-H_0), -x_1(\gamma-\delta x_1)-u_2(x_1,x_2)(H(x_1,x_2)-H_0))$

$\tilde{f}(x) = f(x) - u(x)(H(x)-H_0) = f(x) - k  \nabla{H}(x) (H(x)-H_0)$

Dans le schéma d'Euler, on a:

$H(x^{j+1})-H_0 = H(x^j+dt \tilde{f}(t_j,x^j))-H_0$

$= H(x^j)+dt \nabla{H}(x^j)^T \tilde{f}(t_j,x^j)-H_0+O(dt^2)$

$=H(x^j)+dt \nabla{H}(x^j)^T (f(x^j) - k  \nabla{H}(x^j) (H(x^j)-H_0)) - H_0 + O(dt^2)$ 

Or on a vu que $\nabla{H}(x^j)^T f(x) = 0$ (question 4)

Donc $H(x^{j+1})-H_0 = H(x^j) - k dt \nabla{H}(x^j)^T \nabla{H}(x^j) (H(x^j)-H_0)) - H_0 + O(dt^2)$

$H(x^{j+1})-H_0 = (1-k dt \|H\|) (H(x^j)-H_0) + O(dt^2)$

Pour qu'il y ait bien convergence, il faut que l'écart H(x^{j+1})-H_0$ tende vers 0, donc que la raison de cette suite soit strictement inférieure à 1 en norme.

Il faut donc que $\|1-k dt \|H\| \| <1$

C'est pourquoi, à $dt$ fixé, $k$ ne peut pas être choisi arbitrairement. Il ne doit pas être trop grand, sinon $\|1-k dt \|H\| \| > 1$ et le schéma est divergent. C'est ce qui s'est passé ci-dessus avec $k = 100$
